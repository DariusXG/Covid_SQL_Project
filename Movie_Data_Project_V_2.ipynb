```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example of analysis: Movie Ratings Distribution
plt.figure(figsize=(10,6))
sns.histplot(data=data, x='rating', bins=20, kde=True)
plt.title('Movie Ratings Distribution')
plt.show()
```

```python
# Load your data - Replace 'your_data.csv' with the path to your dataset
df = pd.read_csv('movies.csv')

# Count the number of movies per genre
genre_counts = df['genre'].value_counts()

# Calculate the average score per genre
average_scores = df.groupby('genre')['score'].mean()

# Combine both metrics into a single DataFrame for easier analysis
genre_popularity = pd.DataFrame({'Number_of_movies': genre_counts, 'Average_score': average_scores})

# Display the DataFrame
print(genre_popularity)
```

```python
# Sort genres by the number of movies
genre_popularity = genre_popularity.sort_values(by='Number_of_movies', ascending=True)

fig, ax1 = plt.subplots(figsize=(10,8))  # Adjust the size of the plot

# Horizontal bar plot for number of movies per genre
ax1.barh(genre_popularity.index, genre_popularity['Number_of_movies'], color='b', alpha=0.5)
ax1.set_xlabel('Number of movies')
ax1.set_ylabel('Genre')
ax1.tick_params('x', colors='b')

# Create second x-axis for average scores
ax2 = ax1.twiny()

# Line plot for average scores per genre
ax2.plot(genre_popularity['Average_score'], genre_popularity.index, color='r')
ax2.set_xlabel('Average score', color='r')
ax2.tick_params('x', colors='r')

fig.tight_layout()
plt.show()
```

```python
# Sort genres by the number of movies
genre_popularity = genre_popularity.sort_values(by='Number_of_movies', ascending=True)

fig, ax1 = plt.subplots(figsize=(10,8))  # Adjust the size of the plot

# Horizontal bar plot for number of movies per genre
bars = ax1.barh(genre_popularity.index, genre_popularity['Number_of_movies'], color='b', alpha=0.5)
ax1.set_xlabel('Number of movies')
ax1.set_ylabel('Genre')
ax1.tick_params('x', colors='b')

# Add data points to the bars
for bar in bars:
    ax1.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             str(int(bar.get_width())), va='center')

# Create second x-axis for average scores
ax2 = ax1.twiny()

# Line plot for average scores per genre
ax2.plot(genre_popularity['Average_score'], genre_popularity.index, color='r')
ax2.set_xlabel('Average score', color='r')
ax2.tick_params('x', colors='r')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the movies.csv data into a DataFrame
movies_df = pd.read_csv('movies.csv')

# Create an empty DataFrame to store the highest grossing movies
highest_grossing_movies = pd.DataFrame()

# Iterate over each year in the range
for year in range(1980, 2021):
    # Filter the original DataFrame for the given year and find the movie with the highest gross
    highest_grossing_movie = movies_df[movies_df['year'] == year].nlargest(1, 'gross')
    # Append the highest grossing movie of the year to the DataFrame
    highest_grossing_movies = highest_grossing_movies.append(highest_grossing_movie)

# Plot the highest grossing movies
plt.figure(figsize=(15, 8))
plt.bar(highest_grossing_movies['year'], highest_grossing_movies['gross'])
plt.xlabel('Year')
plt.ylabel('Gross Revenue')
plt.title('Highest Grossing Movies (1980 - 2020)')
plt.show()
```
